# 03 Explore and Viz Deals Data

## Import modules

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Import data

```python
! ls ../../data/interim/
```

```python
path = '../../data/interim/deals.csv'
df = pd.read_csv(path, index_col = 0)
df.head(3)
```

## Visualize firm age by codex category

```python
sns.boxplot(x='codex_category', y='firm_age', data=df)
plt.title('Firm age at funding by Codex Category')
plt.xticks(rotation = 90);
```

# Firm age at funding by deal type

```python
order = list(df.deal_type.value_counts().sort_index().index)
# Modify this to manually change this
order = ['Angel',
         'Convertible Note',
         'Debt Financing',
         'Equity crowdfunding',
         'Grant',
         'Non-equity assistance',
         'Post-IPO Equity',
         'Private Equity',
         'Secondary Market',
         'Seed Fund',
         'Seed fund',
         'Series A',
         'Series B',
         'Series C',
         'Series D',
         'Series E',
         'Series F',
         'Venture Capital']
```

```python
sns.boxplot(x='deal_type', y='firm_age', data=df, order=order)
plt.title('Firm age at funding by deal type')
plt.xticks(rotation = 90);
```

## Show amount of money by deal year

```python
df.groupby(['deal_year'])['round_total'].sum().plot(kind='bar')
plt.title('$ (100M) funded per year');
```

```python
# It would be cool to put a widget here
fig = plt.figure(figsize=(15,5))

df.groupby(['deal_type'])['round_total'].sum().plot(kind='bar')
plt.title('$ (100M) funded by year');
```

```python
df.groupby(['codex_category'])['round_total'].sum().plot(kind='bar')
plt.title('Funding by codex category');
```

## Try Money by Firm Age?

```python
df.groupby(['firm_age'])['round_total'].sum().plot(kind='bar')
plt.title('Funding by codex category');
```

```python
df.groupby(['firm_age'])['round_total'].count().plot(kind='bar')
plt.title('Funding by codex category');
```
