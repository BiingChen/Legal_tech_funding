# 01 Explore and Viz Firms Data

## Import modules

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

## Import data

```python
! ls ../../data/interim/
```

```python
filepath = '../../data/interim/firms.csv'
df = pd.read_csv(filepath, index_col = 0)
df.head(3)
```

## How many companies are we looking at?  Any duplicates?

```python
df.shape
```

We have data on 335 companies.  Each row of data represents a single company.

## Breakdown by Codex Category

```python
df.codex_category.value_counts()
```

```python
codex = df.codex_category.value_counts(normalize=True)
sns.barplot(x=codex.index, y = codex.values, )
plt.title('% of firms by Codex Category')
plt.xticks(rotation=90);
```

## Breakdown for country

```python
country = df.country.value_counts(normalize=False)

sns.barplot(x=country.index, y = country.values, )

plt.title('% of firms by country')
plt.xticks(rotation=90);
```

## Breakdown by city and country (top 10)

```python
city_country = df.groupby(['country','city'])['firm_id'].count().sort_values(ascending=False)
plt.title('Top 10 count of firms by city/country')
city_country.head(10).plot(kind='bar');
```

## Binary Variables:  Market, Analytics, AI, PvI

```python
fig = plt.figure(figsize=(12,3))
fig.add_subplot(141)
df.market.value_counts().plot(kind='bar')
plt.title('Firms market focus')
plt.xticks(rotation=0);

fig.add_subplot(142)
df.analytics.value_counts().plot(kind='bar')
plt.title('Firms utilizing analytics')
plt.xticks([0,1],['Yes','No'], rotation=0);

fig.add_subplot(143)
df.ai.value_counts().plot(kind='bar')
plt.title('Firms utilizing AI')
plt.xticks([0,1],['Yes','No'], rotation=0);

fig.add_subplot(144)
df.point_integrated.value_counts().plot(kind='bar')
plt.title('Integrated vs Point')
plt.xticks([0,1],['Integrated','Point'], rotation=0);


plt.tight_layout()
```

## Examine Problem Space

**Notes:** There are 3 columns associated with problems.  We can cut this up and look at the problems in a number of different ways.  Let's go through them:
1. Only rank 1 problems
2. All listed problems

### Only Rank 1 problems

```python
df.problem_1.value_counts(normalize=True).plot(kind='bar')
plt.title('Rank 1 problems covered by firms');
```

### All problems in any position

```python
problems = list(df.problem_1) + list(df.problem_2) + list(df.problem_3)
problems = [x for x in problems if x !='NONE']
problems = pd.Series(problems)
plt.title('All problems covered by firms')
problems.value_counts(normalize=True).plot(kind='bar');
```
