## First attempt at creating this visual using real data

```python
data = df[['problem_1','problem_2','problem_3']]
```

```python
def add_node(G, node_name):
    if node_name != 'NONE':
        node = pydot.Node(node_name)
        G.add_node(node)
```

```python
G = pydot.Dot(graph_type ='digraph')

for key, row in data.iterrows():
    
    p1 = row.problem_1
    p2 = row.problem_2
    p3 = row.problem_3

    # Add nodes
    add_node(G, p1)
    add_node(G, p2)
    add_node(G, p3)

    # Add edges
    if p1 != 'NONE' and p2 !='NONE':
        edge1_2 = pydot.Edge(p1, p2)
        G.add_edge(edge1_2)

    if p2 != 'NONE' and p3 !='NONE':
        edge2_3 = pydot.Edge(p2, p3)
        G.add_edge(edge2_3)
        
im = Image(G.create_png())
display(im)
```

## Write functions to create the viz

```python
def add_node(G, node_name):
    if node_name != 'NONE':
        node = pydot.Node(node_name)
        G.add_node(node)


color_pallet = []
color_pallet.append(('red','blue','green'))
color_pallet.append(('grey0','grey20','grey40'))
color_pallet.append(('skyblue4','skyblue3','skyblue'))


pal_num = 0
        
# Convert this into a function
def create_viz(data):
    G = pydot.Dot(graph_type ='digraph')

    for key, row in data.iterrows():

        p1 = row.problem_1
        p2 = row.problem_2
        p3 = row.problem_3

        # Add nodes
        if p1 != 'NONE':
            node1 = pydot.Node('1_'+p1, style='filled', color = color_pallet[pal_num][0], fillcolor = 'white')
            G.add_node(node1)
        
        if p2 != 'NONE':
            node2 = pydot.Node('2_'+p2, style='filled', color = color_pallet[pal_num][1], fillcolor = 'white')
            G.add_node(node2)

        if p3 != 'NONE':
            node3 = pydot.Node('3_'+p3, style='filled', color = color_pallet[pal_num][2], fillcolor = 'white')
            G.add_node(node3)
            

        # Add edges
        if p1 != 'NONE' and p2 !='NONE':
            edge1_2 = pydot.Edge(node1, node2)
            G.add_edge(edge1_2)

        if p2 != 'NONE' and p3 !='NONE':
            edge2_3 = pydot.Edge(node2, node3)
            G.add_edge(edge2_3)

    im = Image(G.create_png())
    display(im)
```

## I should just add a widget here

```python
# Create the functions that take the category and then does the viz
def update_viz(category):
    data = df.loc[df.codex_category==category,
              ['codex_category','problem_1','problem_2','problem_3']]
    create_viz(data)
```
